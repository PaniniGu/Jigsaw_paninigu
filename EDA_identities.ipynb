```python
import gc 
import os
import pandas as pd
from tqdm import tqdm
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

tqdm.pandas()

def read_data(fpath):
    train = pd.read_csv(fpath, index_col='id')
    return train

def analysis(df, col_name):
#     df = read_data(fpath)
    
    #分析df['severe_toxicity']
    plt.figure()
    plt.title('hist of ' + col_name)
    plt.hist(df[col_name], bins=100)
    
    print("总样本数：{}".format(len(df)))
    print(col_name + "不为空的样本数：{}".format(df[col_name].count()))
    num1 = len([i for i in df[col_name] if i == 0])
    print(col_name + "值等于0的样本数：{}".format(num1))
    print(col_name + "值大于0的样本数：{}".format(len(df) - num1))
    
    print(col_name + "的最小值：{}".format(df[col_name].min()))
    print(col_name + "的平均值：{}".format(df[col_name].mean()))
    print(col_name + "的中位数：{}".format(df[col_name].median()))
    print(col_name + "的75%分位数：{}".format(np.percentile(df[col_name], 75)))
    print(col_name + "的85%分位数：{}".format(np.percentile(df[col_name], 85)))
    print(col_name + "的最大值：{}".format(df[col_name].max()))
    print(col_name + "的std：{}".format(df[col_name].std()))
    
    #return df
    
train_file = 'input/train.csv'
male = 'male'
female = 'female'
homosexual_gay_or_lesbian = 'homosexual_gay_or_lesbian'
christian = 'christian'
jewish = 'jewish'
muslim = 'muslim'
black = 'black'
white = 'white'
psychiatric_or_mental_illness = 'psychiatric_or_mental_illness'

transgender = 'transgender'
other_gender = 'other_gender'
heterosexual = 'heterosexual'
bisexual = 'bisexual'
other_sexual_orientation = 'other_sexual_orientation'
hindu = 'hindu'
buddhist = 'buddhist'
atheist = 'atheist'
other_religion = 'other_religion'
asian = 'asian'
latino = 'latino'
other_race_or_ethnicity = 'other_race_or_ethnicity'
physical_disability = 'physical_disability'
intellectual_or_learning_disability = 'intellectual_or_learning_disability'
other_disability = 'other_disability'

df = read_data(train_file)
analysis(df, male)
```
